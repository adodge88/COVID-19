```python
import pandas as pd
```

```python
data = "assets/data/covid19.csv"
data_df = pd.read_csv(data)
data_df.head()
```

```python
all_to_date_df = data_df[['country_region', 'confirmed_to_date']]
all_to_date_df.head()
```

```python
country_count_df = all_to_date_df.groupby(['country_region']).sum()
country_count_df.head()
```

```python
country_count_df.to_csv('assets/data/confirmed_country_count.csv')
```
