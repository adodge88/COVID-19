```python
data = "assets/data/covid19.csv"
data_df = pd.read_csv(data)
data_df.head()
```

```python
unique_ids = data_df.id_loc.unique()
print (len(unique_ids))
```

```python
most_recent_data_df = data_df.tail(len(unique_ids))
most_recent_data_df.head()
```

```python
all_to_date_df = most_recent_data_df[['country_region', 'confirmed_to_date','recovered_to_date','deaths_to_date']]
all_to_date_df.head()
```

```python
country_count_df = all_to_date_df.groupby(['country_region']).sum()
country_count_df.head()
```

```python
country_count_df.to_csv('assets/data/country_count.csv')
```
